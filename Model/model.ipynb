```python
import pandas as pd

df = pd.read_csv("../Dataset/water_potability.csv")
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
num_cols = ['ph','Hardness','Solids','Chloramines','Sulfate','Conductivity','Organic_carbon','Trihalomethanes','Turbidity']
X = df[num_cols]
y = df['Potability']
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.2, test_size=0.8, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score, accuracy_score, precision_score, recall_score
from sklearn.pipeline import Pipeline

def get_test_model_score(test_models_list):
    for test_model in test_models_list:
        model_name = test_model[0]
        model = test_model[1]   
        pipeline = Pipeline(steps=[
            ('model', model)
        ])
        pipeline.fit(train_X, train_y)
        pred_y = pipeline.predict(test_X)
        print(f'{model_name} metrics:\n Accuracy score: {accuracy_score(test_y, pred_y)}\n Precision score: {precision_score(test_y, pred_y)}\n Recall score: {recall_score(test_y, pred_y)}\n F1-Score: {f1_score(test_y, pred_y)}\nConfusion Matrix: \n{confusion_matrix(test_y, pred_y)}')
            
test_models = [
    ('LinearSVC', LinearSVC(random_state=1)),
    ('KNeighbors', KNeighborsClassifier()),
    ('RandomForest', RandomForestClassifier(random_state=1)),
    ('GradientBoosting', GradientBoostingClassifier(random_state=1)),
]

get_test_model_score(test_models)
```

```python
search = GridSearchCV(GradientBoostingClassifier(random_state=1), param_grid=[{
    'max_depth': [1, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400, 425, 450, 475, 500],
    'max_leaf_nodes': [2, 25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400, 425, 450, 475, 500]
}], scoring='precision', n_jobs=-1)

search.fit(train_X, train_y)
print(search.best_params_)
```

```python
model = GradientBoostingClassifier(random_state=1, max_depth=25, max_leaf_nodes=50)
model.fit(train_X, train_y)
pred_y = model.predict(test_X)
print(f'Model metrics:\n Accuracy score: {accuracy_score(test_y, pred_y)}\n Precision score: {precision_score(test_y, pred_y)}\n Recall score: {recall_score(test_y, pred_y)}\n F1-Score: {f1_score(test_y, pred_y)}\nConfusion Matrix: \n{confusion_matrix(test_y, pred_y)}')
```

```python
import joblib
best_model = search.best_estimator_
joblib.dump(best_model, 'src_model.pkl')
```
